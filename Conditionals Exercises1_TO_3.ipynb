```python
# Exercises: Level 1
# Level 1 Exercises

# Task 1: Driving Age Feedback
age = int(input("Enter your age: "))
if age >= 18:
    print("You are old enough to learn to drive.")
else:
    print(f"You need {18 - age} more years to learn to drive.")

# Task 2: Comparing Ages
my_age = 25  # Example age
your_age = int(input("Enter your age: "))
if your_age > my_age:
    difference = your_age - my_age
    if difference == 1:
        print("You are 1 year older than me.")
    else:
        print(f"You are {difference} years older than me.")
elif your_age < my_age:
    difference = my_age - your_age
    if difference == 1:
        print("I am 1 year older than you.")
    else:
        print(f"I am {difference} years older than you.")
else:
    print("We are the same age.")

# Task 3: Comparing Two Numbers
a = int(input("Enter number one: "))
b = int(input("Enter number two: "))
if a > b:
    print(f"{a} is greater than {b}")
elif a < b:
    print(f"{a} is smaller than {b}")
else:
    print(f"{a} is equal to {b}")

# Level 2 Exercises

# Task 1: Grading System
score = int(input("Enter your score: "))
if 80 <= score <= 100:
    grade = "A"
elif 70 <= score <= 79:
    grade = "B"
elif 60 <= score <= 69:
    grade = "C"
elif 50 <= score <= 59:
    grade = "D"
else:
    grade = "F"
print(f"Your grade is {grade}.")

# Task 2: Season Checker
month = input("Enter the month: ").capitalize()
if month in ['September', 'October', 'November']:
    print("The season is Autumn.")
elif month in ['December', 'January', 'February']:
    print("The season is Winter.")
elif month in ['March', 'April', 'May']:
    print("The season is Spring.")
elif month in ['June', 'July', 'August']:
    print("The season is Summer.")
else:
    print("Invalid month entered.")

# Task 3: Fruit List Management
fruits = ['banana', 'orange', 'mango', 'lemon']
fruit = input("Enter a fruit: ").lower()
if fruit in fruits:
    print("That fruit already exists in the list.")
else:
    fruits.append(fruit)
    print("Modified list of fruits:", fruits)

# Level 3 Exercises

# Person Dictionary Task
person = {
    'first_name': 'Asabeneh',
    'last_name': 'Yetayeh',
    'age': 250,
    'country': 'Finland',
    'is_marred': True,
    'skills': ['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address': {
        'street': 'Space street',
        'zipcode': '02210'
    }
}

# Check if 'skills' key exists and print the middle skill
if 'skills' in person:
    skills = person['skills']
    middle_skill = skills[len(skills) // 2]
    print("The middle skill is:", middle_skill)

# Check if 'Python' skill exists
if 'skills' in person and 'Python' in person['skills']:
    print("The person has Python skill.")

# Check for developer title
if 'skills' in person:
    skills = person['skills']
    if set(skills) == {'JavaScript', 'React'}:
        print("He is a front end developer.")
    elif {'Node', 'Python', 'MongoDB'}.issubset(skills):
        print("He is a backend developer.")
    elif {'React', 'Node', 'MongoDB'}.issubset(skills):
        print("He is a fullstack developer.")
    else:
        print("Unknown title.")

# Check if person is married and lives in Finland
if person.get('is_marred') and person.get('country') == 'Finland':
    print(f"{person['first_name']} {person['last_name']} lives in Finland. He is married.")
```
